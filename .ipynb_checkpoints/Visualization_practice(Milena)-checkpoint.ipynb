## About
-This is a data on student performance and report on **Math**, **Reading**, and **Writing** based on **Gender**, **Race**, and **Parental level of education**.

-In this data frame the "test preparation course" column seems to be missing more that 60% of its data.

```python
#This is a function for a seaborn scatter plot graph of three series in the dataframe

def stud_perf(col1, col2, col3):
    data_frame_graph = sns.scatterplot(data=df, x=col1, y=col2, hue=col3)
    
    return data_frame_graph
    
```

```python
stud_perf('Math Score', 'Reading Score', 'Gender')
```

Generally the 'female' and 'male' test scores have a strong correlation. According to the plot above the 'female' gender seem to have scored higher in 'Reading' as compared to the 'male' who scored better in 'Math'. We do notice a few outliers in the plot of very low scores in 'Reading' and 'Math'.

```python
df_graph = sns.histplot(df['Gender'])
```

```python
df.Gender.value_counts()
```

The 'female' gender of our data exceeds the 'male' entries in the database we have by 36.

```python
#This is a function to visualize the relationship between two series in the dataframe for Student Performance.

def stud_perf1(col1, col2):
    fig, ax = plt.subplots(figsize=(16,5))
    perf_graph = ax.barh(df[col1], df[col2])
    ax.set_xlabel("{}".format(col2))
    ax.set_ylabel("{}".format(col1))
    ax.set_title("{} vs {}".format(col2, col1))
    
    return perf_graph
 
```

```python
stud_perf1('Parental Level Of Education', 'Writing Score')
```

```python
stud_perf1('Parental Level Of Education', 'Reading Score')
```

As we see in the graph above showing the relationship between the "Parental Level of Education" and the test scores, there seem to be no correlation to show that Parent level of education affects students performance in the test.

```python
stud_perf1('Race/Ethnicity', 'Reading Score')
```

```python
stud_perf1('Race/Ethnicity', 'Writing Score')
```

```python
stud_perf1('Race/Ethnicity', 'Math Score')
```

According to the above plots Race/Ethnicity seem to affect the performance of the Students especially of group A, group B, and group C.
